# programacion orientada a objetos


```python
class Libro:   # class starts with capital letter
    def __init__ (self, titulo, autor, genero):
        self.titulo = titulo
        self.autor = autor
        self.genero = genero
        
    def mostrarInformacion(self):  #method with small letter
        print(f"Titulo: {self.titulo}")
        print(f"Autor: {self.autor}")
        print(f"Genero: {self.genero}")
        
# crear objetos/instancias a partir de la clase - 
ObjetoLibro1 = Libro("el gran gatsby","F. Scott", "Ficcion")   
ObjetoLibro2 = Libro("titanic","Decaprio", "Ficcion") 

# acceder a los attributos de la clase,
print(ObjetoLibro1.autor)

# acceder a la informacion de los objetos a traves del metodo de la clase
ObjetoLibro1.mostrarInformacion()
ObjetoLibro2.mostrarInformacion()

```

```python
class Vehiculo:
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo
        
    def conducir(self):
        print("el vehiculo esta en movimiento")
              
class Automovil(Vehiculo):
    def __init__(self, marca, modelo, color):
              super().__init__(marca,modelo)    #utilizando la super clase
              self.color = color
              
    def abrirPuertas(self):
              print("las puertas estan abiertas")
              
              
auto1 = Automovil("toyota", "camry", "sand")
print(auto1.marca)
print(auto1.modelo)
print(auto1.color)
auto1.conducir()
auto1.abrirPuertas()

              
              
              
            
```

```python
# polimorfismo

# no polimorfismo
class Animal:
    def hablar(self, sonido):
        print(sonido)
        
class Perro:
    def ladrar(self):
        print("guuaau")
        
class Gato:
    def maullar(self):
        print("miaiu")
        
class Vaca:
    def mu(self):
        print("muuuu")
        
firulais = Perro()
firulais.ladrar()


misifus = Gato()
misifus.maullar()
```

```python
# usando polimorfismo - usa el mismo metodo para diferentes clases
class Animal:
    def hablar(self, sonido):
        print(sonido)
        
class Perro(Animal):
    pass
        
class Gato(Animal):
    pass
        
class Vaca(Animal):
    pass
        
firulais = Perro()
firulais.hablar("guuaauuu")


misifus = Gato()
misifus.hablar("mmmiiaauu")
```
