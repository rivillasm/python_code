```python
# method str

class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad
        
    def __str__(self):
        return f"nombre: {self.nombre}, edad:{self.edad}"
    
    def __eq__(self,other):
        return self.nombre == other.nombre and self.edad==other.edad
        
        
persona1 = Persona("mario",42)
persona2 = Persona("carlos",40)
print(persona1==persona2)

persona3 = Persona("mario",42)
print(persona1==persona3)
```

```python
# Decoradores - funciones especiales que permiter modificar el comportamiento de
# otras funciones sin cambiar su codigo interno.

def decorador(func):
    def wrapper():
        suma = 1+1
        print("antes de llamar a la funcion")
        func()
        print("despues de llamar a la funcion")
             
    return wrapper
          
@decorador              
def funcion():
        print("soy la funcion")
              
funcion()
              
```

```python
# tambien se pueden utilizar decoradores en classes
# __call__ en clases hace las veces de un wrapper

class DecoradorClass:
    def __init__(self, func):
        self.func = func
        
    def __call__(self):  
        print("antes de llamar a la funcion")
        self.func()
        print("despues de llamar a la funcion")
        
@DecoradorClass        
def otrafuncion():
    print("hola 7")
    
otrafuncion()
```

### * Recursividad

```python
def calcular_factorial(num):
    factorial=1
    for i in range(1,num+1):
        factorial = factorial*i
        
    return factorial

resultado = calcular_factorial(3)
print(resultado)
```

```python
def factorial_recursivo(num):  # a function calls itself
    if num==0:
        return 1
    else:
        return num*factorial_recursivo(num-1)
    
resultado = factorial_recursivo(3)
print(resultado)
```

### * composicion de funciones 



```python
def suma(n1,n2):
    return n1+n2

def duplicar(num):
    return num*2

def composicion(num1,num2):
    result1=duplicar(num1)
    result2=duplicar(num2)
    return suma(result1,result2)

result = composicion(10,20)
print(result)
```

```python
# usando lambdas

dulicar_lambda = lambda num1:num1*2
suma_lambda = lambda num1,num2:num1+num2

comp_lambda = lambda num1,num2:suma_lambda(dulicar_lambda(num1),dulicar_lambda(num2))
result = comp_lambda(10,20)
print(result)
```

```python
import tkinter as tk
from tkinter import messagebox
```

```python
# agregar tarea a una lista
def agregar_tarea():
    tarea = entrada_tarea.get()  # obtiene el texto del cuadro de la entrada
    if tarea:
        lista_tareas.insert(tk.END, tarea)
        entrada_tarea.delete(0,tk.END)
        
# Eliminar tarea seleccionada de la lista
def eliminar_tarea():
    try: 
        tarea_seleccionada = lista_tareas.curselection()[0]  # indice de tarea seleccinada
        lista_tareas.delete(tarea_seleccionada)
    except IndexError:
        messagebox.showwarning("Advertencia", "porfavor seleccione una tarea")
        
# config ventana principal
ventana = tk.Tk()
ventana.title('***** LISTA DE TAREAS *****')

# cuadro de entrada para agregar tares
entrada_tarea = tk.Entry(ventana, width=40)
entrada_tarea.pack(pady=10)

# boton agregar tarea
boton_agregar = tk.Button(ventana, text='Agregar Tarea', command=agregar_tarea)
boton_agregar.pack(pady=5)


# crear lista tareas
lista_tareas = tk.Listbox(ventana, width=50)
lista_tareas.pack(padx=10, pady=10)

# boton eliminar tareas
boton_eliminar = tk.Button(ventana, text='Eliminar Tarea', command=eliminar_tarea)
boton_eliminar.pack(pady=5)

ventana.mainloop()
```
