```python
def player_guess():
    guess=''  #empty guess
    while guess not in ['0','1','2']:
        guess = input ("Pick a number: 0, 1 or 2")  #input alwasy output a string
    
    return int(guess)
    
```

```python
def check_guess(mylist,guess):
    if mylist[guess] == '0':
        print("correct!")
    else:
        print("wrong guess!")
        print(mylist)
        
    
    
```

```python
def lesser_of_two_events(a,b):
    if (a%2==0 and b%2==0):
        return min(a,b)
    else:
        return max(a,b)
    
print(lesser_of_two_events(2,4))
print(lesser_of_two_events(2,5))
```

```python
def old_mcdonald(name):
    new_word = [ch.upper() if (i==0 or i==3) else ch for i,ch in enumerate(name)]
    return "".join(new_word)
    
print(old_mcdonald('macdonald'))
```

```python
def has_33(nums):
    """find two consecutives numbers 3, thus 33,  if it does exist"""

    for j in range(len(nums)-1):
        if nums[j]==nums[j+1]:
            return True
    return False

nums = [1,3,3]    
print(has_33(nums))
nums = [1,3,1,3]    
print(has_33(nums))
nums = [3,1,3]    
print(has_33(nums))
nums = [1,3,1,1,3,3]    
print(has_33(nums))
```

```python
def blackjack(num1,num2,num3):
    if num1>11 or num2>11 or num3>11:
        return "Wrong number, is higher than 11"
    
    total = num1+num2+num3
    
    if total <= 21:
        return total
    elif (num1==11 or num2==11 or num3==11):
        return total-10
    else:
        return 'BUST'
            

print(blackjack(5,6,7))
print(blackjack(9,9,9))
print(blackjack(9,9,11))
```

```python
def summer_69(arr):
    num1, num2 = 6, 9
    if not ((num1 in arr) and (num2 in arr)):
        return sum(arr)
    else:
        index_num1 = arr.index(num1)
        index_num2  = arr.index(num2)
        if(len(arr)==index_num2):
            return sum(arr[:index_num1])
        else:
            return sum(arr[:index_num1]+arr[index_num2+1:])
        
    
    
print(summer_69([1,3,5]))
print(summer_69([4,5,6,7,8,9]))
print(summer_69([2,1,6,9,11]))
print(summer_69([1,6,9]))
```

```python
def count_primes(num):
    #check for 0 and 1
    if num < 2:
        return 0
    
    #initialize the list of primes
    primes=[2]
    x=3
    
    while x<=num:
        for y in range(3,x,2):
            if x%y==0:
                x+=2
                break
        primes.append(x)
        x+=2
    
    print(primes)
    return len(primes)
                

count_primes(100)
    
```

```python
def splicer(mystring):
    if len(mystring)%2==0:
        return 'Even'
    else:
        return mystring[0]
        
names = ['mario', 'jess', 'jasonn', 'dilka']
list(map(splicer, names))
```

```python
def check_even(num):
    return num%2 == 0

mynums = [1,2,3,4,5,6]

list(filter(check_even,mynums))
```

```python
x=50
def func():   #notice x was not given as a parameter
    
    # by declaring global, any changes of x inside the func will affect x globally
    global x   
    print(f'x is {x}')
    
    #local reassignment on a global bariable
    x=200
    print(f'locally changed globally x to {x}')
```

```python
func()
print(x)
```

```python
import string


def up_low(s:str)->str:
    upper=0
    lower=0
    new_s:str = s.translate(str.maketrans('', '', string.punctuation)).replace(' ','')
    for character in new_s:
        if character.isupper():
            upper+=1
        else:
            lower+=1
    return upper,lower    
    
s = 'Hello Mr. Rogers, how are you this fine Tuesday?'
up, low = up_low(s)
f'there are {up} upper characters and {low} lower characters'
```

```python
import string


def up_low(s:str)->str:
    new_s:str = s.translate(str.maketrans('', '', string.punctuation)).replace(' ','')
    upper = list(filter(lambda x: x.isupper(), new_s))
    return len(upper), len(new_s)-len(upper)    
    
s = 'Hello Mr. Rogers, how are you this fine Tuesday?'
up, low = up_low(s)
f'there are {up} upper characters and {low} lower characters'
```
