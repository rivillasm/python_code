```python
def player_guess():
    guess=''  #empty guess
    while guess not in ['0','1','2']:
        guess = input ("Pick a number: 0, 1 or 2")  #input alwasy output a string
    
    return int(guess)
    
```

```python
player_guess()
```

```python
myindex = player_guess()
myindex
```

```python
def check_guess(mylist,guess):
    if mylist[guess] == '0':
        print("correct!")
    else:
        print("wrong guess!")
        print(mylist)
        
    
    
```

```python
# initial list
mylist = ['', '0','']

#shuffle list
mixed_list = shuffle_list(mylist)

# user guess
guess = player_guess()

# check guess
check_guess(mixed_list,guess)
```

```python
def myfunc(**kwargs):
    print(kwargs)
    if 'fruit' in kwargs:   # returns a dictionary
        print("my fruit of choise is {}".format(kwargs['fruit']))
    else:
        print('I did not find any fruit here')
```

```python
myfunc(fruit = 'apple', veggie = 'letturce')
```

```python
def myfunc(*args, **kwargs):
    print(args)
    print(kwargs)
    print('I would like {} {}'.format(args[0], kwargs['food']))
    
myfunc(10,20,30,fruit='orange', food='eggs', animal='dog')

# Note keep the order in you call firts *args then you provide first those arguments and only 
# then you add the kwargs items
```

```python
def lesser_of_two_events(a,b):
    if (a%2==0 and b%2==0):
        return min(a,b)
    else:
        return max(a,b)
    
print(lesser_of_two_events(2,4))
print(lesser_of_two_events(2,5))
```

```python
def animal_crackers(text):
    
    if(text.lower().split()[0][0]==text.lower().split()[1][0]):
        return True
    else:
        return False

print(animal_crackers("Levelheaded Llama"))
print(animal_crackers("Levelheaded llama"))
print(animal_crackers("Crazy Kangaroo"))
```

```python
def old_mcdonald(name):
    new_word = [ch.upper() if (i==0 or i==3) else ch for i,ch in enumerate(name)]
    return "".join(new_word)
    
print(old_mcdonald('macdonald'))
```

```python
def has_33(nums):
    """find two consecutives numbers 3, thus 33,  if it does exist"""

    for j in range(len(nums)-1):
        if nums[j]==nums[j+1]:
            return True
    return False

nums = [1,3,3]    
print(has_33(nums))
nums = [1,3,1,3]    
print(has_33(nums))
nums = [3,1,3]    
print(has_33(nums))
nums = [1,3,1,1,3,3]    
print(has_33(nums))
```

```python
def blackjack(num1,num2,num3):
    if num1>11 or num2>11 or num3>11:
        return "Wrong number, is higher than 11"
    
    total = num1+num2+num3
    
    if total <= 21:
        return total
    elif (num1==11 or num2==11 or num3==11):
        return total-10
    else:
        return 'BUST'
            

print(blackjack(5,6,7))
print(blackjack(9,9,9))
print(blackjack(9,9,11))
```

```python
def summer_69(arr):
    num1, num2 = 6, 9
    if not ((num1 in arr) and (num2 in arr)):
        return sum(arr)
    else:
        index_num1 = arr.index(num1)
        index_num2  = arr.index(num2)
        if(len(arr)==index_num2):
            return sum(arr[:index_num1])
        else:
            return sum(arr[:index_num1]+arr[index_num2+1:])
        
    
    
print(summer_69([1,3,5]))
print(summer_69([4,5,6,7,8,9]))
print(summer_69([2,1,6,9,11]))
print(summer_69([1,6,9]))
```

```python
def count_primes(num):
    #check for 0 and 1
    if num < 2:
        return 0
    
    #initialize the list of primes
    primes=[2]
    x=3
    
    while x<=num:
        for y in range(3,x,2):
            if x%y==0:
                x+=2
                break
        primes.append(x)
        x+=2
    
    print(primes)
    return len(primes)
                

count_primes(100)
    
```
