```python
# Write a function that takes in a board and a mark (X or O) 
# and then checks to see if that mark has won

def win_check(board, mark)->bool:
    row_seq:int = 0
    rows:int = 3  #expected 3x3 matrix
    columns:int = 3  #expected 3x3 matrix
    
    #check for wining along the rows
    for i in range(0,len(board),rows): 
        row_seq+=1
        if len(set(board[i:i+columns]))==1:
            print(f'the winner player is {mark} at row {row_seq}')
            return True
        
    #check for wining along the columns    
    for j in range(0,columns): 
        if len(set(board[j::rows]))==1:
            print(f'the winner player is {mark} at column {j+1}')
            return True
        
    #check for wining diagonals
    if (len(set(board[0:8:4]))==1 or len(set(board[2:6:2]))==1):
        print(f'the winner player is {mark} in diagonal')
        return True
    
    return False
    
    
```

```python
def space_check(board, position)->bool:
    """ check if position is available == empty in this case"""
    return board[position] == " "
```

```python
def full_board_check(board)->bool:
    """ check and return True if all the spaces in the board are full"""
    for i in range(len(board)):
        if space_check(board,i):  # return false if there is an empty space
            return False
    return True  # if board is full
    
```

```python
def player_choice(board):
    """ input a position and check if the position is empty"""
    position:int = 0
    positions_all = ["1","2","3","4","5","6","7","8","9"]
    while position not in positions_all or not space_check(board,position):
        position = int(input("Choose a position: (1-9) "))
        if position not in positions_all:
            print("Sorry, Please Choose a position: (1-9) to continue...")
    
    return position
```

```python
def replay()->bool:
    play:str = ' '
    while play not in ["Y","N"]:
        play = input("Do you want to replay?, Choose (Y or N): ").upper()
        if play not in ["Y","N"]:
            print("Sorry, Please enter 'Y' or 'N' to continue...")
    
    if play == "Y":
        return True
    else:
        return False
    
```
